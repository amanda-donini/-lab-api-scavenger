```python
if response.status_code == 200:
    forks = response.json()
    languages = set() 
    
  
    for fork in forks:
        languages_url = fork['languages_url']
    
        languages_response = requests.get(languages_url)
        
        if languages_response.status_code == 200:
            fork_languages = languages_response.json()
            languages.update(fork_languages.keys())
        else:
            print(f"Failed to fetch languages for a fork. Status code: {languages_response.status_code}")
    
    language_list = list(languages)
    
    print(language_list)
else:
    print(f"Failed to fetch forks. Status code: {response.status_code}")


```

```python
three_weeks_ago = (datetime.now() - timedelta(weeks=3)).isoformat()

# API endpoint to list commits for the repository since the calculated date
api_url = f"https://api.github.com/repos/ta-data-pt-rmt/-lab-api-scavenger/commits?since={three_weeks_ago}"

# Send a GET request to the API
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    commits = response.json()
    num_commits = len(commits)
    
    # Print the number of commits made in the past 3 weeks
    print(f"Number of commits made in the past 3 weeks: {num_commits}")
else:
    print(f"Failed to fetch commits. Status code: {response.status_code}")
```
